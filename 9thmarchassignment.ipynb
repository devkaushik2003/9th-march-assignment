Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

The normal distribution, also known as the Gaussian distribution, is a widely used probability distribution that is commonly used as a model in various situations

The parameters of the normal distribution are the mean (μ) and the standard deviation (σ). The mean represents the center of the distribution, while the standard deviation represents the spread or variability of the distribution.

The mean (μ) determines the location of the peak of the normal distribution, and the shape of the distribution is symmetric with respect to the mean. A higher mean shifts the distribution to the right, while a lower mean shifts it to the left.

The standard deviation (σ) determines the spread or variability of the distribution. A larger standard deviation results in a wider distribution, while a smaller standard deviation results in a narrower distribution.

In summary, the parameters of the normal distribution, namely the mean and standard deviation, play a crucial role in determining the shape, location, and spread of the distribution, and they are important in interpreting and using the normal distribution as a model in various applications.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a binary outcome, where there are only two possible outcomes, typically denoted as "success" or "failure

The binomial distribution, on the other hand, models the number of successes in a fixed number of independent and identically distributed Bernoulli trials. In other words, it represents the number of successes in a sequence of binary trials, where each trial has the same probability of success, denoted as p.

An example of a situation where the Bernoulli distribution can be used is modeling the outcome of a single coin flip, where "success" can represent getting heads (or tails, depending on the convention chosen), and "failure" can represent getting tails (or heads).

The main differences between the Bernoulli distribution and the binomial distribution are:

Number of trials: The Bernoulli distribution has only one trial, while the binomial distribution has a fixed number of trials, denoted as n.

Number of successes: The Bernoulli distribution models a single success or failure outcome, while the binomial distribution models the number of successes in n trials.

Probability of success: In the Bernoulli distribution, the probability of success (p) is a fixed parameter, whereas in the binomial distribution, the probability of success (p) is the same for each trial, but the number of trials (n) is fixed.
